**Lista de dados disponíveis nos dados abertos da Câmara dos Deputados:**

**1 - Blocos**
- **/blocos** - Lista de dados sobre os blocos partidários
- **/blocos/{id}** - Informações sobre um bloco partidário específico

**2 - Deputados**
- **/deputados** - Listagem e busca de deputados, segundo critérios
- **/deputados/{id}** - Informações detalhadas sobre um deputado específico
- **/deputados/{id}/despesas** - As despesas com exercício parlamentar do deputado
- **/deputados/{id}/discursos** - Os discursos feitos por um deputado em eventos diersos
- **/deputados/{id}/eventos** - Uma lista de eventos com a participação do parlamentar
- **/deputados/{id}/frentes** - As frentes parlamentares das quais um deputado é integrante
- **/deputados/{id}/ocupacoes** - Os empregos e atividades que o(a) deputado(a) já teve
- **/deputados/{id}/orgaos** - Os órgãos dos quais um deputado é integrante
- **/deputados/{id}/profissoes** - As profissões que o parlamentar declarou à Câmara que já exerceu ou que pode exercer pela sua formação e/ou experiência
- **/legislaturas/{id}/mesa** - Quais deputados fizeram parte da Mesa Diretora em uma legislatura
- **/referencias/deputados/codSituacao** - As possíveis situações de exercício parlamentar de um deputado
- **/referencias/situacoesDeputado** - As possíveis situações de exercício parlamentar de um deputado

**3 - Eventos**
- **/eventos** - Lista de eventos ocorridos ou previstos nos diversos órgãos da Câmara
- **/eventos/{id}** - Informações detalhadas sobre um evento específico
- **/eventos/{id}/deputados** - Os deputados participantes de um evento específico
- **/eventos/{id}/orgaos** - Lista de órgãos organizadores do evento
- **/eventos/{id}/pauta** - Lista de proposições que foram ou deverão ser analisadas em um evento de caráter deliberativo
- **/eventos/{id}/votacoes** - Informações detalhadas de votações sobre um evento específico
- **/referencias/eventos/codSituacaoEvento** - As possíveis situações para eventos
- **referencias/eventos/codTipoEvento** - Os tipos de eventos realizados na Câmara
- **referencias/situacoesEvento** - As possíveis situações para eventos
- **referencias/tiposEvento** - Os tipos de eventos realizados na Câmara

**4 - Frentes**
- **/frentes** - Lista de frentes parlamentares de uma ou mais legislaturas
- **/frentes/{id}** - Informações detalhadas sobre uma frente parlamentar
- **/frentes/{id}/membros** - Os deputados que participam de uma frente parlamentar

**5 - Legislaturas**
- **/legislaturas** - Os períodos de mandatos e atividades parlamentares da Câmara
- **/legislaturas/{id}** - Informações extras sobre uma determinada legislatura da Câmara
- **/legislaturas/{id}/mesa** - Quais deputados fizeram parte da Mesa Diretora em uma legislatura

**6 - Partidos**
- **/partidos** - Os partidos políticos que têm ou já tiveram parlamentares em exercício na Câmara
- **/partidos/{id}** - Informações detalhadas sobre um partido
- **/partidos/{id}/membros** - Uma lista dos parlamentares de um partido durante um período

**7 - Proposições**
- **/proposicoes** - Lista configurável de proposições na Câmara
- **/proposicoes/{id}** - Informações detalhadas sobre uma proposição específica
- **/proposicoes/{id}/autores** - Lista pessoas e/ou entidades autoras de uma proposição
- **/proposicoes/{id}/relacionadas** - Uma lista de proposições relacionadas a uma em especial
- **/proposicoes/{id}/temas** - Lista de áreas temáticas de uma proposiçao
- **/proposicoes/{id}/votacoes** - Informações detalhadas de votações sobre uma proposição específica
- **/referencias/proposicoes/codSituacao** - Os possíveis estados de tramitação de uma proposição
- **/referencias/proposicoes/codTema** - Os vários tipos de temas existentes
- **/referencias/proposicoes/codTipoAutor** - Entidades que podem ser autoras de proposições
- **/referencias/proposicoes/codTipoTramitacao** - Os vários tipos de tramitação existentes
- **/referencias/proposicoes/siglaTipo** - Os vários tipos de proposições existentes
- **/referencias/situacoesProposicao** - Os possíveis estados de tramitação de uma proposição
- **/referencias/tiposAutor** - Entidades que podem ser autoras de proposições
- **/referencias/tiposProposicao** - Os vários tipos de proposições existentes
- **/referencias/tiposTramitacao** - Os vários tipos de tramitação existentes

**8 - Referências**
- **/referencias/deputados** - Valores válidos para parâmetros do endpoint '/deputados'
- **/referencias/deputados/codSituacao** - As possíveis situações do exercício parlamentar de um deputado
- **/referencias/deputados/codTipoProfissao** - Códigos e títulos de atividades profissionais usados na Câmara
- **/referencias/deputados/siglaUF** - As siglas e nomes dos estados e do Distrito Federal
- **/referencias/deputados/tipoDespesa** - Os possíveis tipos de despesas que fazem parte da Cota Parlamentar
- **/referencias/eventos** - Valores váidos para parâmetros do endpoint '/eventos'
- **/referencias/eventos/codSituacaoEvento** - As possíveis situações para evento
- **/referencias/eventos/codTipoEvento** - Os tipos de eventos realizados na Câmara
- **/referencias/orgaos** - Valores válidos para parâmetros do endpoint sobre órgãos
- **/referencias/orgaos/codSituacao** - As situações em que órgãos podem se encontrar
- **/referencias/orgaos/codTipoOrgao** - Os tipos de órgãos que existem na Câmara
- **/referencias/orgaos/proposicoes** - Valores válidos para parâmetros do endpoint sobre órgãos
- **/referencias/proposicoes/codSituacao** - Os possíveis estados de tramitação de uma proposição
- **/referencias/proposicoes/codTema** - Os vários tipos de temas existentes
- **/referencias/proposicoes/codTipoAutor** - Entidades que podem ser autoras de proposições
- **/referencias/proposicoes/codTipoTramitacao** - Os vários tipos de tramitação existentes
- **/referencias/proposicoes/siglaTipo** - Os vários tipos de proposições existentes
- **/referencias/situacoesDeputado** - As possíveis situações de exercício parlamentar de um deputado
- **/referencias/situacoesEvento** - As possíveis situações para eventos
- **/referencias/situacoesOrgao** - As situações em que órgãos podem se encontrar
- **/referencias/situacoesProposicao** - Os possíveis estados de tramitação de uma proposição
- **/referencias/tiposAutor** - Entidades que podem ser autoras de proposições
- **referencias/tiposEvento** - Os tipos de eventos realizados na Câmara
- **/referencias/tiposOrgao** - Os tipos de órgãos que existem na Câmara
- **/referencias/tiposProposicao** - Os vários tipos de proposições existentes
- **/referencias/tiposTramitacao** - Os vários tipos de tramitação existentes
- **/referencias/uf** - As siglas e nomes dos estados e do Distrito Federal

**9 - Votações**
- **/votacoes** - Lista de votações da Câmara
- **/votacoes/{id}** - Informações detalhadas sobre uma votação da Câmara
- **/votacoes/{id}/orientacoes** - O voto recomendado pelas lideranças aos seus deputados em uma votação
- **/votacoes/{id}/votos** - Como cada parlamentar votou em uma votação nominal e aberta

**10 - Órgãos**
- **/orgaos** - A lista das comissões e outros órgãos legislativos da Câmara
- **/orgaos/{id}** - Informações detalhadas sobre um órgão da Câmara
- **/orgaos/{id}/eventos** - Os eventos ocorridos ou previstos em um órgão legislativo
- **/orgaos/{id}/membros** - Lista de cargos de um órgão e parlamentas que os ocupam
- **/orgaos/{id}/votacoes** - Informações detalhadas sobre votações de um órgão da Câmara
- **referencias/orgaos/codSituacao** - As situações em que órgãos podem se encontrar
- **referencias/situacoesOrgao** - As situações em que órgãos podem se encontrar
